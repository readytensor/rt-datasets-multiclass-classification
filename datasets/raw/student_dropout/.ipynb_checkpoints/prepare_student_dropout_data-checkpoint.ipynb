```python
import numpy as np
import pandas as pd
import os, sys
```

```python
dataset_name = 'student_dropout'
inp_fname = 'data.csv'
```

```python
input_dir = './data'
output_dir = f'./../../processed/{dataset_name}/'
outp_fname = os.path.join(output_dir, f'{dataset_name}.csv')
```

# Read Data

```python
data = pd.read_csv(os.path.join(input_dir, inp_fname), sep=";")
data.head()
```

```python
id_col = "Id"
target_col = "Target"
```

```python
data[target_col].value_counts()
```

# Insert Id Column

```python
# insert Id column 
if id_col not in data.columns:
    N = data.shape[0]
    data.insert(0, id_col, np.arange(N))
    print(data.head())
data[id_col] = data[id_col].astype(str)
```

# Save Main Data File

```python
data.to_csv(outp_fname, index=False)
```
