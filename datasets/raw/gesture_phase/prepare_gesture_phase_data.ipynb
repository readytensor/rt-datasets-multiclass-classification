```python
import numpy as np
import pandas as pd
import os, sys
```

```python
dataset_name = 'gesture_phase'
inp_fname = 'gesture_phase.csv'
```

```python
input_dir = './data'
output_dir = f'./../../processed/{dataset_name}/'
outp_fname = os.path.join(output_dir, f'{dataset_name}.csv')
```

# Read Data

```python
cols = [f"X{i}" for i in range(1, 33)] + ['Phase']
cols   
```

```python
len(set(cols))
```

```python
data = pd.read_csv(os.path.join(input_dir, inp_fname), header=None, names=cols)
data.head()
```

```python
id_col = "Id"
target_col = "Phase"
```

# Prepare Data

```python
data['Phase'] = data['Phase'].str.replace("'", "")
```

```python
data.head()
```

```python
data['Phase'].value_counts()
```

# Insert Id Column

```python
# insert Id column 
if id_col not in data.columns:
    N = data.shape[0]
    data.insert(0, id_col, np.arange(N))
    print(data.head())
data[id_col] = data[id_col].astype(str)
```

# Save Main Data File

```python
data.to_csv(outp_fname, index=False)
```
