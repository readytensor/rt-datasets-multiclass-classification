```python
import numpy as np
import pandas as pd
import os
```

```python
dataset = 'statlog'
inp_fname1 = 'sat.trn'
inp_fname2 = 'sat.test'
```

```python
input_dir = './raw'
output_dir = './processed'

outp_fname = f'{dataset}.csv'
outp_train_fname = f'{dataset}_train.csv'
outp_test_fname = f'{dataset}_test.csv'

output_dir_train = os.path.join(output_dir, 'train')
output_dir_test = os.path.join(output_dir, 'test')
```

# Read Data

```python
cols = [
    "age",
    "sex",
    "histologic-type",
    "degree-of-diffe",
    "bone",
    "bone-marrow",
    "lung",
    "pleura",
    "peritoneum",
    "liver",
    "brain",
    "skin",
    "neck",
    "supraclavicular",
    "axillar",
    "mediastinum",
    "abdominal",
    "class",
]
```

```python
data = pd.read_csv(os.path.join(input_dir, inp_fname1)
#                    , header=None, names=cols
                  )
data.head()
```

```python
data.shape
```

# Insert Id Column

```python
# insert Id column 
N = data.shape[0]
data.insert(0, 'id', np.arange(N))
data.head()
```

# Shuffle Data

```python
# shuffle data
data = data.sample(frac=1, random_state=42)
data.head()
```

# Replace null indicators

```python
data = data.replace('?', np.nan)
data.head()
```

# Update Class labels

mapping = {
    "+": "positive", 
    "-": "negative"
}

data['class'] = data['class'].map(mapping)
data.head()

# Save Main Data File

```python
data.to_csv(os.path.join(output_dir, outp_fname), index=False)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
test_size = 0.1

data_train, data_test = train_test_split(data, test_size=test_size, random_state=42)
print(data_train.shape, data_test.shape)

data_train.to_csv(os.path.join(output_dir_train, outp_train_fname), index=False)
data_test.to_csv(os.path.join(output_dir_test, outp_test_fname), index=False)
```
